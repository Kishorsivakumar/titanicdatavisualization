```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("C:\\Users\\Kishor\\Downloads\\titanic\\titanic_dataset.csv")

```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.describe()

```

```python
df['Sex'].value_counts()
```

```python
ages = df['Age'].values
print(np.mean(ages), np.std(ages))
print(ages[ages > 60])  # Senior passengers
```

```python

plt.hist(df['Age'], bins=20, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("People Surived")
plt.grid()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Surived by Gender")
plt.grid()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Boxplot by Pclass")
```

```python

df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric = df_numeric.fillna(df_numeric.mean())
plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='Blues')
plt.title("Correlation Matrix")

```

```python
g = sns.FacetGrid(df, col='Survived')
g.map(plt.hist, 'Age', bins=20)
```
